## Histogramas de distancia al robot

```python
plt.figure(figsize=(12,6))
counts, edges, bars = plt.hist(robot_10a['distancia_pelota_10a'],edgecolor="black")
plt.bar_label(bars)
plt.title("Histograma de DISTANCIA A PELOTA de 10A (en centimetros)")
plt.show()
```

```python
plt.figure(figsize=(12,6))
counts, edges, bars = plt.hist(robot_10b['distancia_pelota_10b'],edgecolor="black")
plt.bar_label(bars)
plt.title("Histograma de DISTANCIA A PELOTA de 10B (en centimetros)")
plt.show()
```

```python
plt.figure(figsize=(12,6))
counts, edges, bars = plt.hist(robot_17b['distancia_pelota_17b'],edgecolor="black")
plt.bar_label(bars)
plt.title("Histograma de DISTANCIA A PELOTA de 17B (en centimetros)")
plt.show()
```
